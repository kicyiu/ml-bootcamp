```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost-revenue-clean.csv')
```

```python
data.describe()
```

```python
5.034000e+03
```

```python
X = DataFrame(data, columns = ['production_budget_usd'])
y = DataFrame(data, columns = ['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwider Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X, y)
```

Slope Coefficient:

```python
regression.coef_ # theta_1
```

```python
# Intercept or theta_1 constant

regression.intercept_
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.3) # scatter for draw points

plt.plot(X, regression.predict(X), color='red', linewidth=4) # plot for draw lines

plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwider Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
regression.score(X, y)
```
